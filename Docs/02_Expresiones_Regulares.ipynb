## ğŸ’¡ Mejores PrÃ¡cticas

### âœ… DO - Hacer

1. **Usar lÃ­mites de palabra** (`\b`) para evitar coincidencias parciales
2. **Normalizar a minÃºsculas** antes de aplicar regex
3. **Soportar variantes** (con/sin acentos, singular/plural)
4. **Respuestas variadas** con `random.choice()` para naturalidad
5. **Patrones simples**: Solo para casos obvios y frecuentes

### âŒ DON'T - No hacer

1. **No usar regex para consultas complejas** â†’ Usar RAG/LLM
2. **No patrones demasiado amplios** â†’ Muchos falsos positivos
3. **No patrones demasiado restrictivos** â†’ Muchos falsos negativos
4. **No hardcodear demasiadas variantes** â†’ DifÃ­cil de mantener
5. **No regex para lÃ³gica de negocio** â†’ Solo saludos/despedidas bÃ¡sicos
